## 3. Turning the background of the images to white

```python
import cv2
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt
from tqdm import tqdm

change_background_mp = mp.solutions.selfie_segmentation

change_bg_segment = change_background_mp.SelfieSegmentation()

path = "/Users/parth/Documents/Rutgers Coursework/Semester 3/MDS/DL Project/MDS Project 2/Dataset/Train"
dir_list = os.listdir(path)

for i in tqdm(range(len(dir_list))):
    inside_path = path + "/" + dir_list[i]
    inside_dir_list = os.listdir(inside_path)
    for j in range(len(inside_dir_list)):
        final_path = inside_path + "/" + inside_dir_list[j]

        sample_img = cv2.imread(final_path)

        RGB_sample_img = cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB)

        result = change_bg_segment.process(RGB_sample_img)

        binary_mask = result.segmentation_mask > 0.8

        binary_mask_3 = np.dstack((binary_mask,binary_mask,binary_mask))

        output_image = np.where(binary_mask_3, sample_img, 255) 

        cv2.imwrite("/Users/parth/Documents/Rutgers Coursework/Semester 3/MDS/DL Project/MDS Project 2/Dataset/out/" + dir_list[i] + "/" + inside_dir_list[j], output_image)
```
